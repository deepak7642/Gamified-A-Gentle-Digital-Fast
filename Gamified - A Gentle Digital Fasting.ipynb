```python
# Select features for clustering
features = df[["HoursSpent",
    "Distraction", "Restlessness", "Anxiety", "ConcentrationDifficulty",
    "SelfComparision", "ValidationSeeking", "Depression",
    "Sleeplessness", "ActivityInterestVariance"]]

# -----------------------------
# Scale features â€” very important for K-Means
# -----------------------------
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
# -----------------------------
# Find optimal k with Elbow Method
# -----------------------------
inertia = []
k_range = range(2, 10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_scaled)
    inertia.append(kmeans.inertia_)  # Distortion / SSE

import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=list(k_range),
        y=inertia,
        mode='lines+markers',
        marker=dict(color='royalblue', size=8),
        line=dict(width=2),
        name='Inertia'
    )
)

fig.update_layout(
    title='Elbow Method For Optimal k',
    xaxis_title='Number of clusters (k)',
    yaxis_title='Inertia (SSE)',
    xaxis=dict(tickmode='linear'),
    template='plotly_white',
    width=800,
    height=500
)

fig.show()
```

```python
# Suppose you choose k= from the elbow
optimal_k = 5

kmeans_final = KMeans(n_clusters=optimal_k, random_state=42)
df['Social_Impact_Type'] = kmeans_final.fit_predict(features_scaled)

# Select columns by passing a list of column names
print(df.groupby('Social_Impact_Type')[features.columns].mean())
```

```python
# =========================
# Gamified Digital Wellness Recommender
# =========================

import pandas as pd
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from txtai.pipeline import LLM as TxtaiLLM
from langchain_core.language_models.llms import BaseLLM as LangChainLLM
from langchain_core.runnables import RunnableSequence

# =========================
# Wrap txtai LLM for LangChain
# =========================
txtai_llm_instance = TxtaiLLM("muranAI/gemma-3n-E4B-it-GGUF/gemma-3n-e4b-it-q2_k.gguf")

class TxtaiLangChainLLM(LangChainLLM):
    def _call(self, prompt, stop=None):
        return txtai_llm_instance(prompt)

    @property
    def _llm_type(self):
        return "txtai"

llm = TxtaiLangChainLLM()

# =========================
# Helper: Map HoursSpent â†’ Buckets
# =========================
def map_hours_to_bucket(hours):
    """
    Convert numeric HoursSpent into exact rule-based buckets.
    """
    try:
        hours = float(hours)
        if hours > 5:
            return "More than 5"
        elif 2 <= hours <= 5:
            return "2 to 5"
        else:
            return "Less than 2"
    except:
        # In case HoursSpent is already categorical string
        val = str(hours).strip()
        if val in ["More than 5", "2 to 5", "Less than 2"]:
            return val
        else:
            return "Unknown"

# =========================
# Recommendation Prompt (Wellness + Gamified Nudges)
# =========================
recommend_action_prompt = PromptTemplate.from_template("""
You are a digital wellness mentor who gives gamified nudges and challenges.

RULES:
1. If Predicted_Social_Impact_Segmt == 'Critical Risk':
   â†’ âš ï¸ Level: SOS Guardian â†’ Challenge: Start with a "1-hour no-screen" mission today, then connect with a supportive friend or family member.

2. If Predicted_Social_Impact_Segmt == 'High Risk':
   â†’ âš–ï¸ Level: Balance Builder â†’ Challenge: Try 2 offline activities today (walk, journaling, outdoor play) and track how you feel.

3. If Predicted_Social_Impact_Segmt == 'Moderate Risk':
   â†’ âš’ï¸ Level: Habit Shaper â†’ Challenge: Set a 2-hour screen-time cap and swap one digital activity with a hobby.

4. If Predicted_Social_Impact_Segmt == 'Mild Risk':
   â†’ ðŸŒ± Level: Growth Seeker â†’ Challenge: Share 1 positive post, call a family member, or join a community activity.

5. If Predicted_Social_Impact_Segmt == 'Healthy':
   â†’ ðŸ† Level: Wellness Champion â†’ Challenge: Celebrate your balance streak, log your digital habits, and mentor a peer with your success tips.

HOURSPENT LEVELS:
- If HoursSpent == 'More than 5':
   â†’ ðŸ”´ Time Overlord â†’ Extra Challenge: Take a 5-min break every hour and spend at least 30 mins outdoors.

- If HoursSpent == '2 to 5':
   â†’ ðŸ”² Balance Keeper â†’ Bonus: Maintain your healthy range by mixing short offline breaks with creative hobbies.

- If HoursSpent == 'Less than 2':
   â†’ ðŸŸ¢ Focus Master â†’ Bonus: Youâ€™re on a great streak! Keep celebrating your strong balance and inspire others.

OUTPUT FORMAT:
Return ONLY the matched recommendation lines.
- Do NOT repeat or restate the rules.
- Do NOT include conditions or "If..." statements.
- Only output the selected challenge(s).


INPUT:
- Social Impact Segment: {Predicted_Social_Impact_Segmt}
- HoursSpent: {HoursSpent}""")


recommend_chain = recommend_action_prompt | llm | StrOutputParser()

# =========================
# Short Explanation Prompt
# =========================
llm_explain_prompt = PromptTemplate.from_template("""
You are a friendly digital well-being mentor.

User Info:
- Social Impact Segment: {Predicted_Social_Impact_Segmt}
- HoursSpent: {HoursSpent}

System's recommended action:
{action}

In 1â€“2 upbeat sentences (max 30 words), explain why this challenge fits the user.
Keep it encouraging and gamified.

Answer:
""")

llm_explain_chain = llm_explain_prompt | llm | StrOutputParser()

# =========================
# Run recommender for X_test with User Input
# =========================
def run_recommender_for_user(X_test_df):
    try:
        # Ask user to input User_ID
        user_id = int(input("Enter User_ID: "))

        row = X_test_df.loc[X_test_df["User_ID"] == user_id]

        if row.empty:
            return f"âŒ User_ID {user_id} not found in dataset."

        row = row.iloc[0]
        seg = str(row["Predicted_Social_Impact_Segmt"]).strip()
        hrs_bucket = map_hours_to_bucket(row["HoursSpent"])

        # Generate challenge
        action = recommend_chain.invoke({
            "Predicted_Social_Impact_Segmt": seg,
            "HoursSpent": hrs_bucket
        })

        # Generate short explanation
        explanation = llm_explain_chain.invoke({
            "Predicted_Social_Impact_Segmt": seg,
            "HoursSpent": hrs_bucket,
            "action": action
        })

        return {
            "User_ID": row["User_ID"],
            "Segment": seg,
            "HoursSpent": hrs_bucket,
            "Challenge": action,
            "Why": explanation
        }

    except ValueError:
        return "âŒ Invalid input. Please enter a numeric User_ID."


# =========================
# Example usage
# =========================
if __name__ == "__main__":
    result = run_recommender_for_user(X_test_df) # Use X_test_df here
    if isinstance(result, dict):
        print("\nðŸŽ¯ Personalized Recommendation")
        print(f"User_ID: {result['User_ID']}")
        print(f"Segment: {result['Segment']}")
        print(f"HoursSpent: {result['HoursSpent']}")
        print(f"Challenge: {result['Challenge']}")
        print(f"Why: {result['Why']}")
    else:
        print(result)
```

```python
import gradio as gr
import pandas as pd
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from txtai.pipeline import LLM as TxtaiLLM
from langchain_core.language_models.llms import BaseLLM as LangChainLLM
from langchain_core.outputs import Generation, LLMResult

# =========================
# txtai â†’ LangChain Wrapper
# =========================
txtai_llm_instance = TxtaiLLM(
    "muranAI/gemma-3n-E4B-it-GGUF/gemma-3n-e4b-it-q2_k.gguf"
)

class TxtaiLangChainLLM(LangChainLLM):
    def _call(self, prompt, stop=None):
        return txtai_llm_instance(prompt)

    def _generate(self, prompts, stop=None, **kwargs) -> LLMResult:
        generations = []
        for prompt in prompts:
            text = self._call(prompt, stop=stop)
            generations.append([Generation(text=text)])
        return LLMResult(generations=generations)

    @property
    def _llm_type(self):
        return "txtai"

llm = TxtaiLangChainLLM()

# =========================
# Helper
# =========================
def map_hours_to_bucket(hours):
    try:
        hours = float(hours)
        if hours > 5:
            return "More than 5"
        elif 2 <= hours <= 5:
            return "2 to 5"
        else:
            return "Less than 2"
    except:
        return str(hours)

# =========================
# Prompts
# =========================
recommend_action_prompt = PromptTemplate.from_template("""
You are a digital wellness mentor who gives gamified nudges and challenges.

RULES:
1. If Predicted_Social_Impact_Segmt == 'Critical Risk':
   â†’ âš ï¸ Level: SOS Guardian â†’ Challenge: Start with a "1-hour no-screen" mission today, then connect with a supportive friend or family member.

2. If Predicted_Social_Impact_Segmt == 'High Risk':
   â†’ âš–ï¸ Level: Balance Builder â†’ Challenge: Try 2 offline activities today (walk, journaling, outdoor play).

3. If Predicted_Social_Impact_Segmt == 'Moderate Risk':
   â†’ âš’ï¸ Level: Habit Shaper â†’ Challenge: Set a 2-hour screen-time cap and swap one digital activity with a hobby.

4. If Predicted_Social_Impact_Segmt == 'Mild Risk':
   â†’ ðŸŒ± Level: Growth Seeker â†’ Challenge: Share 1 positive post or connect with family.

5. If Predicted_Social_Impact_Segmt == 'Healthy':
   â†’ ðŸ† Level: Wellness Champion â†’ Challenge: Celebrate your balance streak and mentor a peer.

HOURSPENT LEVELS:
- More than 5 â†’ ðŸ”´ Time Overlord
- 2 to 5 â†’ ðŸ”² Balance Keeper
- Less than 2 â†’ ðŸŸ¢ Focus Master

Return ONLY the matched recommendation lines.

INPUT:
- Social Impact Segment: {Predicted_Social_Impact_Segmt}
- HoursSpent: {HoursSpent}
""")

recommend_chain = recommend_action_prompt | llm | StrOutputParser()

llm_explain_prompt = PromptTemplate.from_template("""
In 1â€“2 upbeat sentences (max 30 words), explain why this challenge fits the user.

User:
- Segment: {Predicted_Social_Impact_Segmt}
- HoursSpent: {HoursSpent}

Challenge:
{action}
""")

llm_explain_chain = llm_explain_prompt | llm | StrOutputParser()

# =========================
# Gradio Callback Function
# =========================
def fetch_user_info(user_id):
    row = X_test_df[X_test_df["User_ID"] == user_id]

    if row.empty:
        return "User not found", "", ""

    row = row.iloc[0]
    seg = str(row["Predicted_Social_Impact_Segmt"])
    hrs = map_hours_to_bucket(row["HoursSpent"])

    action = recommend_chain.invoke({
        "Predicted_Social_Impact_Segmt": seg,
        "HoursSpent": hrs
    })

    explanation = llm_explain_chain.invoke({
        "Predicted_Social_Impact_Segmt": seg,
        "HoursSpent": hrs,
        "action": action
    })

    user_info = f"""
User ID: {row['User_ID']}
Segment: {seg}
Hours Spent: {hrs}
"""

    return user_info.strip(), action.strip(), explanation.strip()

# =========================
# Gradio UI
# =========================
with gr.Blocks() as demo:
    gr.Markdown("# ðŸŒðŸ’¡ Gamified Digital Wellness Recommender")

    with gr.Row():
        with gr.Column(scale=1):
            user_id = gr.Dropdown(
                choices=sorted(X_test_df["User_ID"].unique().tolist()),
                label="Select User ID"
            )
            fetch_btn = gr.Button("Get Recommendation ðŸŽ¯")

        with gr.Column(scale=2):
            user_info = gr.Textbox(label="User Info", lines=4)
            system_action = gr.Textbox(label="Challenge", lines=4)
            llm_explanation = gr.Textbox(label="Why This Challenge?", lines=3)

    fetch_btn.click(
        fetch_user_info,
        inputs=user_id,
        outputs=[user_info, system_action, llm_explanation]
    )

demo.launch()
```
