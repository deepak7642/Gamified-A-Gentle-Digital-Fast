```python
# Select features for clustering
features = df[["HoursSpent",
    "Distraction", "Restlessness", "Anxiety", "ConcentrationDifficulty",
    "SelfComparision", "ValidationSeeking", "Depression",
    "Sleeplessness", "ActivityInterestVariance"]]

# -----------------------------
# Scale features — very important for K-Means
# -----------------------------
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
# -----------------------------
# Find optimal k with Elbow Method
# -----------------------------
inertia = []
k_range = range(2, 10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(features_scaled)
    inertia.append(kmeans.inertia_)  # Distortion / SSE

import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=list(k_range),
        y=inertia,
        mode='lines+markers',
        marker=dict(color='royalblue', size=8),
        line=dict(width=2),
        name='Inertia'
    )
)

fig.update_layout(
    title='Elbow Method For Optimal k',
    xaxis_title='Number of clusters (k)',
    yaxis_title='Inertia (SSE)',
    xaxis=dict(tickmode='linear'),
    template='plotly_white',
    width=800,
    height=500
)

fig.show()
```

```python
# Suppose you choose k= from the elbow
optimal_k = 5

kmeans_final = KMeans(n_clusters=optimal_k, random_state=42)
df['Social_Impact_Type'] = kmeans_final.fit_predict(features_scaled)

# Select columns by passing a list of column names
print(df.groupby('Social_Impact_Type')[features.columns].mean())
```

```python
# =========================
# Gamified Digital Wellness Recommender
# =========================

import pandas as pd
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.schema import StrOutputParser
from txtai.pipeline import LLM as TxtaiLLM
from langchain.llms.base import LLM as LangChainLLM

# =========================
# Wrap txtai LLM for LangChain
# =========================
txtai_llm_instance = TxtaiLLM("muranAI/gemma-3n-E4B-it-GGUF/gemma-3n-e4b-it-q2_k.gguf")

class TxtaiLangChainLLM(LangChainLLM):
    def _call(self, prompt, stop=None):
        return txtai_llm_instance(prompt)

    @property
    def _llm_type(self):
        return "txtai"

llm = TxtaiLangChainLLM()

# =========================
# Helper: Map HoursSpent → Buckets
# =========================
def map_hours_to_bucket(hours):
    """
    Convert numeric HoursSpent into exact rule-based buckets.
    """
    try:
        hours = float(hours)
        if hours > 5:
            return "More than 5"
        elif 2 <= hours <= 5:
            return "2 to 5"
        else:
            return "Less than 2"
    except:
        # In case HoursSpent is already categorical string
        val = str(hours).strip()
        if val in ["More than 5", "2 to 5", "Less than 2"]:
            return val
        else:
            return "Unknown"

# =========================
# Recommendation Prompt (Wellness + Gamified Nudges)
# =========================
recommend_action_prompt = PromptTemplate.from_template("""
You are a digital wellness mentor who gives gamified nudges and challenges.

RULES:
1. If Predicted_Social_Impact_Segmt == 'Critical Risk':
   → 🚨 Level: SOS Guardian → Challenge: Start with a "1-hour no-screen" mission today, then connect with a supportive friend or family member.

2. If Predicted_Social_Impact_Segmt == 'High Risk':
   → ⚖️ Level: Balance Builder → Challenge: Try 2 offline activities today (walk, journaling, outdoor play) and track how you feel.

3. If Predicted_Social_Impact_Segmt == 'Moderate Risk':
   → 🛠️ Level: Habit Shaper → Challenge: Set a 2-hour screen-time cap and swap one digital activity with a hobby.

4. If Predicted_Social_Impact_Segmt == 'Mild Risk':
   → 🌱 Level: Growth Seeker → Challenge: Share 1 positive post, call a family member, or join a community activity.

5. If Predicted_Social_Impact_Segmt == 'Healthy':
   → 🏆 Level: Wellness Champion → Challenge: Celebrate your balance streak, log your digital habits, and mentor a peer with your success tips.

HOURSPENT LEVELS:
- If HoursSpent == 'More than 5':
   → 🔴 Time Overlord → Extra Challenge: Take a 5-min break every hour and spend at least 30 mins outdoors.

- If HoursSpent == '2 to 5':
   → 🟡 Balance Keeper → Bonus: Maintain your healthy range by mixing short offline breaks with creative hobbies.

- If HoursSpent == 'Less than 2':
   → 🟢 Focus Master → Bonus: You’re on a great streak! Keep celebrating your strong balance and inspire others.

OUTPUT FORMAT:
Return ONLY the matched recommendation lines.
- Do NOT repeat or restate the rules.
- Do NOT include conditions or "If..." statements.
- Only output the selected challenge(s).


INPUT:
- Social Impact Segment: {Predicted_Social_Impact_Segmt}
- HoursSpent: {HoursSpent}
""")


recommend_chain = LLMChain(
    llm=llm,
    prompt=recommend_action_prompt,
    output_parser=StrOutputParser()
)

# =========================
# Short Explanation Prompt
# =========================
llm_explain_prompt = PromptTemplate.from_template("""
You are a friendly digital well-being mentor.

User Info:
- Social Impact Segment: {Predicted_Social_Impact_Segmt}
- HoursSpent: {HoursSpent}

System's recommended action:
{action}

In 1–2 upbeat sentences (max 30 words), explain why this challenge fits the user.
Keep it encouraging and gamified.

Answer:
""")

llm_explain_chain = LLMChain(
    llm=llm,
    prompt=llm_explain_prompt,
    output_parser=StrOutputParser()
)

# =========================
# Run recommender for X_test with User Input
# =========================
def run_recommender_for_user(X_test_df):
    try:
        # Ask user to input User_ID
        user_id = int(input("Enter User_ID: "))

        row = X_test_df.loc[X_test["User_ID"] == user_id]

        if row.empty:
            return f"❌ User_ID {user_id} not found in dataset."

        row = row.iloc[0]
        seg = str(row["Predicted_Social_Impact_Segmt"]).strip()
        hrs_bucket = map_hours_to_bucket(row["HoursSpent"])

        # Generate challenge
        action = recommend_chain.run({
            "Predicted_Social_Impact_Segmt": seg,
            "HoursSpent": hrs_bucket
        })

        # Generate short explanation
        explanation = llm_explain_chain.run({
            "Predicted_Social_Impact_Segmt": seg,
            "HoursSpent": hrs_bucket,
            "action": action
        })

        return {
            "User_ID": row["User_ID"],
            "Segment": seg,
            "HoursSpent": hrs_bucket,
            "Challenge": action,
            "Why": explanation
        }

    except ValueError:
        return "❌ Invalid input. Please enter a numeric User_ID."


# =========================
# Example usage
# =========================
if __name__ == "__main__":
    result = run_recommender_for_user(X_test)
    if isinstance(result, dict):
        print("\n🎯 Personalized Recommendation")
        print(f"User_ID: {result['User_ID']}")
        print(f"Segment: {result['Segment']}")
        print(f"HoursSpent: {result['HoursSpent']}")
        print(f"Challenge: {result['Challenge']}")
        print(f"Why: {result['Why']}")
    else:
        print(result)
```

```python
import gradio as gr
import pandas as pd
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.schema import StrOutputParser
from txtai.pipeline import LLM as TxtaiLLM
from langchain.llms.base import LLM as LangChainLLM



# =========================
# Wrap txtai LLM
# =========================
txtai_llm_instance = TxtaiLLM("muranAI/gemma-3n-E4B-it-GGUF/gemma-3n-e4b-it-q2_k.gguf")

class TxtaiLangChainLLM(LangChainLLM):
    def _call(self, prompt, stop=None):
        return txtai_llm_instance(prompt)

    @property
    def _llm_type(self):
        return "txtai"

llm = TxtaiLangChainLLM()

# =========================
# Recommendation Prompt
# =========================
recommend_action_prompt = PromptTemplate.from_template("""
You are a digital wellness mentor who gives gamified nudges and challenges.

RULES:
1. If Predicted_Social_Impact_Segmt == 'Critical Risk':
   → 🚨 Level: SOS Guardian → Challenge: Start with a "1-hour no-screen" mission today, then connect with a supportive friend or family member.

2. If Predicted_Social_Impact_Segmt == 'High Risk':
   → ⚖️ Level: Balance Builder → Challenge: Try 2 offline activities today (walk, journaling, outdoor play) and track how you feel.

3. If Predicted_Social_Impact_Segmt == 'Moderate Risk':
   → 🛠️ Level: Habit Shaper → Challenge: Set a 2-hour screen-time cap and swap one digital activity with a hobby.

4. If Predicted_Social_Impact_Segmt == 'Mild Risk':
   → 🌱 Level: Growth Seeker → Challenge: Share 1 positive post, call a family member, or join a community activity.

5. If Predicted_Social_Impact_Segmt == 'Healthy':
   → 🏆 Level: Wellness Champion → Challenge: Celebrate your balance streak, log your digital habits, and mentor a peer with your success tips.

HOURSPENT LEVELS:
- If HoursSpent == 'More than 5':
   → 🔴 Time Overlord → Extra Challenge: Take a 5-min break every hour and spend at least 30 mins outdoors.

- If HoursSpent == '2 to 5':
   → 🟡 Balance Keeper → Bonus: Maintain your healthy range by mixing short offline breaks with creative hobbies.

- If HoursSpent == 'Less than 2':
   → 🟢 Focus Master → Bonus: You’re on a great streak! Keep celebrating your strong balance and inspire others.

OUTPUT FORMAT:
Return ONLY the matched recommendation lines.
- Do NOT repeat or restate the rules.
- Do NOT include conditions or "If..." statements.
- Only output the selected challenge(s).


INPUT:
- Social Impact Segment: {Predicted_Social_Impact_Segmt}
- HoursSpent: {HoursSpent}
""")

recommend_chain = LLMChain(
    llm=llm,
    prompt=recommend_action_prompt,
    output_parser=StrOutputParser()
)

# =========================
# Explanation Prompt
# =========================
llm_explain_prompt = PromptTemplate.from_template("""
You are a friendly digital well-being mentor.

User Info:
- Social Impact Segment: {Predicted_Social_Impact_Segmt}
- HoursSpent: {HoursSpent}

System's recommended action:
{action}

In 1–2 upbeat sentences (max 30 words), explain why this challenge fits the user.
Keep it encouraging and gamified.

Answer:
""")

llm_explain_chain = LLMChain(
    llm=llm,
    prompt=llm_explain_prompt,
    output_parser=StrOutputParser()
)

# =========================
# Backend Logic
# =========================
def fetch_user_info(user_id):
    row = X_test_df[X_test_df['User_ID'] == int(user_id)]
    if row.empty:
        return "❌ User not found", "", ""

    row = row.iloc[0]
    seg = str(row["Predicted_Social_Impact_Segmt"]).strip()
    hrs_bucket = str(row["HoursSpent"]).strip()

    # Generate recommendation
    action = recommend_chain.run({
        "Predicted_Social_Impact_Segmt": seg,
        "HoursSpent": hrs_bucket
    })

    # Generate explanation
    explanation = llm_explain_chain.run({
        "Predicted_Social_Impact_Segmt": seg,
        "HoursSpent": hrs_bucket,
        "action": action
    })

    user_info = f"User ID: {row['User_ID']}\nSegment: {seg}\nHoursSpent: {hrs_bucket}"
    return user_info, action, explanation

# =========================
# Gradio UI
# =========================
with gr.Blocks() as demo:
    gr.Markdown("# 🌍💡 Gamified Digital Wellness Recommender")

    with gr.Row():
        with gr.Column(scale=1):
            user_id = gr.Dropdown(
                choices=sorted(X_test_df["User_ID"].unique().tolist()),
                label="Select User ID"
            )
            fetch_btn = gr.Button("Get Recommendation")

        with gr.Column(scale=2):
            user_info = gr.Textbox(label="User Info", lines=3)
            system_action = gr.Textbox(label="Challenge", lines=3)
            llm_explanation = gr.Textbox(label="Why This Challenge?", lines=3)

    fetch_btn.click(
        fetch_user_info,
        inputs=[user_id],
        outputs=[user_info, system_action, llm_explanation]
    )

demo.launch()
```
